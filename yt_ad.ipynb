# **OUTLIER DETECTION**

```python
#LIKES
plt.figure(figsize=(8,5))
sns.violinplot(x=df['likes'], inner=None, color='skyblue')

lower = df['likes'].quantile(0.01)
upper = df['likes'].quantile(0.99)
outliers = df[(df['likes'] < lower) | (df['likes'] > upper)]['likes']

sns.scatterplot(x=outliers, y=[0]*len(outliers), color='red', s=10)
plt.title("Distribution + Outliers of Likes")
plt.show()

print(f"Likes - Outliers count: {len(outliers)}")
print(f"Lower bound (1%): {lower}, Upper bound (99%): {upper}\n")
```

                                                             Number of likes per video.

Violin Plot Use: Shows how many videos have a certain number of likes. Positive skew indicates few videos get very high likes.


Outlier Points: Red points show videos with exceptionally high or low likes, indicating high engagement content or poor engagement videos.

PREPROCESSING 

```python
df.duplicated().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Missing values check
df.isnull().sum()
```

```python
# Fill missing with mean (numerical columns)
df['likes'] = df['likes'].fillna(df['likes'].mean())
df['comments'] = df['comments'].fillna(df['comments'].mean())
df['watch_time_minutes'] = df['watch_time_minutes'].fillna(df['watch_time_minutes'].mean())
```

```python
df.isnull().sum()
```

```python
# Drop video_id â€“ not useful for modeling
df.drop(columns=['video_id'], inplace=True)
```

```python
# Convert to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# Extract useful features
df['upload_month'] = df['date'].dt.month
df['upload_dayofweek'] = df['date'].dt.dayofweek
df['upload_day'] = df['date'].dt.day
```

```python
# Drop original date column
df.drop(columns=['date'], inplace=True)
```

Linear Regression cannot handle raw dates, but numeric features extracted from date can capture trends or seasonal effects.

# **Feature Engineering**


Create a new feature â€“ Engagement Rate

This feature shows how engaged viewers are relative to total views.

```python
# Engagement rate = (likes + comments) / views
df['engagement_rate'] = (df['likes'] + df['comments']) / df['views']
df[['views', 'likes', 'comments', 'engagement_rate']].head()
```

```python
# Watch ratio = total watch time / video length
df['watch_ratio'] = df['watch_time_minutes'] / df['video_length_minutes']

# Quick check
df[['watch_time_minutes', 'video_length_minutes', 'watch_ratio']].head()
```

```python
from sklearn.preprocessing import StandardScaler
```

StandardScaler ensures all numeric features are on the same scale for Linear Regression

```python
numeric_cols = ['views', 'likes', 'comments', 'watch_time_minutes', 
                'video_length_minutes', 'subscribers']

scale_cols = numeric_cols + ['engagement_rate', 'watch_ratio']
```

```python
scaler = StandardScaler()
df[scale_cols] = scaler.fit_transform(df[scale_cols])
```
